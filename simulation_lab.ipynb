```python
import numpy as np
import matplotlib.pyplot as p
q1=-3e-9
q2=7e-9
k=9e9
pq1=np.array([0.2,0.3,0])
pq2=np.array([-0.3,0.1,0])
p1=np.array([-0.1,-0.2,0])
p2=np.array([0,0.4,0])
r1=p1-pq1
r1mod=np.linalg.norm(r1)
r1cap=r1/r1mod
E1=((k*q1)/(np.square(r1mod)))*(r1cap)
r2=p1-pq2
r2mod=np.linalg.norm(r2)
r2cap=r2/r2mod
E2=((k*q2)/(np.square(r2mod)))*(r2cap)
print("E1 =",E1)
print("E2 =",E2)
print("E =",E1+E2)
```

```python
import numpy as np
import matplotlib.pyplot as p
q1=-3e-9
q2=7e-9
k=9e9
x=float(input("X component of point = "))
y=float(input("Y component of point = "))
z=float(input("Z component of point = "))
pq1=np.array([0.2,0.3,0])
pq2=np.array([-0.3,0.1,0])
p1=np.array([x,y,z])
r1=p1-pq1
r1mod=np.linalg.norm(r1)
r1cap=r1/r1mod
E1=((k*q1)/(np.square(r1mod)))*(r1cap)
r2=p1-pq2
r2mod=np.linalg.norm(r2)
r2cap=r2/r2mod
E2=((k*q2)/(np.square(r2mod)))*(r2cap)
print("E1 =",E1)
print("E2 =",E2)
print("E =",E1+E2)
magE=np.linalg.norm(E1+E2)
print("Magnitude of E = ",magE)
```
