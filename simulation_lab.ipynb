```python
import matplotlib.pyplot as p
import numpy as np
Hr=int(input("Enter how many harmonics do u want: "))
l=int(input("Enter length of the box: "))
N=np.linspace(1,Hr, Hr)
for n in N:
  x=np.linspace(0,l, 1000)
  y=((np.sqrt(2/l)*np.sin((n*np.pi*x)/l))+n)
  p.plot(x,y)



```
