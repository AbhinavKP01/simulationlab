```python
import numpy as np

def gauss_jordan(A, b):
    n = len(A)
    # Augment the coefficient matrix A with the constant vector b
    augmented_matrix = np.hstack((A, b.reshape(-1, 1)))

    for i in range(n):
        # Find the pivot row and swap if necessary
        pivot_row = np.argmax(np.abs(augmented_matrix[i:, i])) + i
        augmented_matrix[[i, pivot_row]] = augmented_matrix[[pivot_row, i]]

        # Make the diagonal element 1
        pivot = augmented_matrix[i, i]
        augmented_matrix[i,-i] /= pivot

        # Eliminate other elements in the current column
        for j in range(n):
            if j != i:
                augmented_matrix[j] -= augmented_matrix[j, i] * augmented_matrix[i]

    # Extract the solution
    solution = augmented_matrix[:, n]

    return solution

# Example system of equations
A = np.array([[2, 1, -1],
              [-3, -1, 2],
              [-2, 1, 2]])
b = np.array([8, -11, -3])

solution = gauss_jordan(A, b)
print("Solution:", solution)
```

```python
import numpy as np

A = np.array([[2, 1, -1],
              [-3, -1, 2],
              [-2, 1, 2]])
b = np.array([8, -11, -3])

# Augment the coefficient matrix A with the constant vector b
augmented_matrix = np.hstack((A, b.reshape(-1, 1)))

n = len(A)

for i in range(n):
    # Divide the current row by the diagonal element to make it 1
    diagonal_element = augmented_matrix[i, i]
    augmented_matrix[i] /= diagonal_element

    # Eliminate non-diagonal elements in the current column
    for j in range(n):
        if j != i:
            coefficient = augmented_matrix[j, i]
            augmented_matrix[j] -= coefficient * augmented_matrix[i]

# Extract the solution from the augmented matrix
solution = augmented_matrix[:, n]

print("Solution:", solution)

```
