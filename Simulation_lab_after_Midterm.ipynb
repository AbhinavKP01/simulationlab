```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 1.0  # Spring constant
mass = 1.0  # Mass of the particle
hbar = 1.0  # Reduced Planck's constant
N = 501 # Number of positions
xmax = 5.0  # Maximum position
stepsize = xmax / N
time_steps = 1000  # Number of time steps
dt = 0.01  # Time step size

# Initial values
x0 = 0.0  # Initial position
v0 = 0.1  # Initial velocity
psi0 = np.exp(-(x0*2) / 2.0) / np.pi*0.25  # Initial wave function for the ground state
dpsi0 = -x0 * psi0  # Initial derivative

# Initialize arrays
x_values = np.linspace(-xmax, xmax, N)
psi = np.zeros(N, dtype=complex)
dpsi = np.zeros(N, dtype=complex)
potential = 0.5 * k * x_values**2

# Lists to store results for even and odd parity
even_psi = [np.abs(psi0)]
odd_psi = [0]

# Time evolution using the Runge-Kutta method
for t in range(time_steps):
    # Initialize Runge-Kutta variables
    k1 = dpsi
    p1 = -2 * (np.conj(psi) * psi * (0.5 * k * x_values*2 - 0.5 * k * x0*2))
    k2 = dpsi + (0.5 * p1 * dt)
    p2 = -2 * (np.conj(psi + 0.5 * k1 * dt) * (psi + 0.5 * k1 * dt) * (0.5 * k * x_values*2 - 0.5 * k * x0*2))
    k3 = dpsi + (0.5 * p2 * dt)
    p3 = -2 * (np.conj(psi + 0.5 * k2 * dt) * (psi + 0.5 * k2 * dt) * (0.5 * k * x_values*2 - 0.5 * k * x0*2))
    k4 = dpsi + (p3 * dt)
    p4 = -2 * (np.conj(psi + k3 * dt) * (psi + k3 * dt) * (0.5 * k * x_values*2 - 0.5 * k * x0*2))

    # Update psi and dpsi
    psi += (k1 + 2 * k2 + 2 * k3 + k4) * dt / 6
    dpsi += (p1 + 2 * p2 + 2 * p3 + p4) * dt / 6

    # Normalize the wave function (optional)
    psi /= np.sqrt(np.trapz(np.abs(psi)**2, x_values))

    # Append the values for even and odd parity
    if t % 2 == 0:
      even_psi.append(np.abs(psi[int(N/2)]))
    else:
      odd_psi.append(np.abs(psi[int(N/2)]))

# Plot the results
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.plot(x_values, potential, label='Potential Energy')
plt.xlabel('Position (x)')
plt.ylabel('Energy')
plt.legend()

plt.subplot(132)
plt.plot(x_values, even_psi, label='Even Parity')
plt.xlabel('Position (x)')
plt.ylabel('|Psi|')
plt.legend()

plt.subplot(133)
plt.plot(x_values, odd_psi, label='Odd Parity')
plt.xlabel('Position (x)')
plt.ylabel('|Psi|')
plt.legend()

plt.tight_layout()
plt.show()
```
